What are the ten highest profit orders?

```python
# get data from SQL
df = %sql SELECT * from Records
# set df as dataframe
df = df.DataFrame()
# sort values in df
df_top10 = df.sort_values(by=['TotalProfit'], ascending=False)
# reset index 
df_top10 = df_top10.reset_index() 
# call top ten values 
df_top10.head(10)
```

What are the ten lowest profit orders?

```python
# call bottom ten values 
df_top10.tail(10)
```

Create a search that will prompt for an order ID and return all the fields for that record

```python
lis_ID = df_top10['OrderID']
    
#Using the try-else structure, we will display the inputted Order ID Number if it exists in one of the dictionaries
def main():
    #We will ask the user to input an Order ID Number to see if it exists within the 10 records with the highest total profits
    orderID = int(input("Please enter an Order ID Number to locate it: "))
    try:
        index = int(df_top10[df_top10['OrderID'] == orderID].index.values.astype(int))
        print(index)
        

        for i in lis_ID:
            if i == orderID:
                return(df_top10.iloc[index])

    #If the Order ID Number the user inputted is not found in one or both of the dictionaries we created, we will inform the user of this
    except ValueError:
        print("Invarid characters")
main()
```
